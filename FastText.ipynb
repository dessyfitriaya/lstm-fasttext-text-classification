```python
import os
import pandas as pd
from tqdm.auto import tqdm
from nltk.tokenize import word_tokenize
from gensim.models import FastText
```

### load data

```python
df = pd.read_excel('/portofolio-ml/nlp-pemilu2024/bismillah-dataset.xlsx')
df
```

```python
import ast

# untuk mengubah the string representation of list jadi list beneran
df['stemming'] = df['stemming'].apply(ast.literal_eval)
df['string'] = df['stemming'].apply(lambda x: ' '.join(x))
df
```

```python
sentences = [word_tokenize(string) for string in tqdm(df.string)]
sentences[:5]
```

### train fasttext model

```python
model = FastText(sentences, vector_size=128, window=5, min_count=3, epochs=1000, sg=0, hs=0)
```

### save

```python
model.save("model/w2v/testing_fasttext")
```

### load

```python
model = FastText.load("model/w2v/testing_fasttext")
```

### model info

```python
ft = model.wv
```

```python
ft.index_to_key
```

### sanity check

```python
ft.similar_by_word("anies", topn=5)
```

higher order visualization

```python
# install umap-learn jgn umap
# klo terlanjur install umap, uninstall

import umap.umap_ as umap
import numpy as np
import plotly.express as px
```

```python
reducer = umap.UMAP()
embedding = reducer.fit_transform(ft.vectors)
```

```python
df = pd.DataFrame(embedding, columns=["umap1", "umap2"])
df["text"] = ft.index_to_key
```

```python
df
```

```python
fig = px.scatter(df, x="umap1", y="umap2", text="text")
fig.update_traces(textposition='top center')
fig.update_layout(
    height=800,
    title_text='Reduced FastText Visualization'
)
fig.show()
```
